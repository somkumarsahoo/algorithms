#### Sum of numbers from list to a target

```python
def addition_path_to_target(nums: list[int] , target: int  = 0):
    def backtrack_sum(start: int, target: int, nums: list[int] , path: list[int] = []):
        if target==0:
            result.append(path)
            return
        if target<0:
            return
                    
        for i in range(start, len(nums)):
            backtrack_sum(start=i+1,target=target-nums[i], nums=nums, path = path + [nums[i]])
        
        
        
    result = []
        
    backtrack_sum(start=0, target=target, nums=nums , path= [])

    return result
            
```

#### Generate Subset

```python
def generate_subset(nums:list[int]):
    def backtrack_list(nums:list[int], currIdx: int, subsets:list[list[int]], currSubset:list[int]=[]):
        
        if currIdx == len(nums):
            subsets.append(currSubset.copy())
            return

        # If we want to insert nums[currIdx] into currSubset
        currSubset.append(nums[currIdx])                     
        backtrack_list(nums,currIdx+1,subsets, currSubset)

        # If we don't want to insert nums[currIdx] into currSubset
        currSubset.pop()
        backtrack_list(nums,currIdx+1,subsets, currSubset)
        
            
    subsets = []  

    backtrack_list(nums, 0, subsets)

    return subsets
```

```python
def generate_subset2(nums:list[int]):
    def backtrack_list(currIdx: int):
        
        if currIdx == len(nums):
            subsets.append(currSubset.copy())
            return

        # If we want to insert nums[currIdx] into currSubset
        currSubset.append(nums[currIdx])                     
        backtrack_list(currIdx+1)

        # If we don't want to insert nums[currIdx] into currSubset
        currSubset.pop()
        backtrack_list(currIdx+1)
        
    
    subsets = []
    currSubset = []

    backtrack_list(0)

    return subsets
```

#### Number of permutations

```python
def get_permutations(nums:list[int])->list[list[int]]:

    def backtrack():
        if len(currPermutation) == len(nums):
            permutations.append(currPermutation.copy())
            return
        
        # What number should we place in the current permutation list?
        
        for num in nums:
            if not num in usedValues:
                currPermutation.append(num)
                usedValues.append(num)
                backtrack()

                currPermutation.pop()
                usedValues.remove(num)
                    
    permutations = []
    currPermutation = []
    usedValues = []
    backtrack()
    
    return permutations
```

#### Number of permutations with repeated elements

```python
def get_permutations_repeatation(nums: list[int]) -> list[list[int]]:

    def backtrack():
        if len(currPermutation) == len(nums):
            permutations.append(currPermutation.copy())
            return

        # Determine which number should be placed in the current permutation list
        for num in num_count:
            if num_count[num] > 0:
                currPermutation.append(num)
                num_count[num] -= 1
                backtrack()

                currPermutation.pop()
                num_count[num] += 1

    permutations = []
    currPermutation = []
    
    # Count the occurrences of each number using a dictionary
    num_count = {}
    for num in nums:
        num_count[num] = num_count.get(num, 0) + 1
    
    backtrack()

    return permutations

# Test the function
print(get_permutations_repeatation([1, 2, 1]))
```
