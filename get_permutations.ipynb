#### Number of permutations

```python
def get_permutations(nums:list[int])->list[list[int]]:

    def backtrack():
        if len(currPermutation) == len(nums):
            permutations.append(currPermutation.copy())
            return
        
        # What number should we place in the current permutation list?
        
        for num in nums:
            if not num in usedValues:
                currPermutation.append(num)
                usedValues.append(num)
                backtrack()

                currPermutation.pop()
                usedValues.remove(num)
                    
    permutations = []
    currPermutation = []
    usedValues = []
    backtrack()
    
    return permutations
```

```python
get_permutations([1,2,3]) # The above logic works with repetation of elemets in the list
```

```python
get_permutations([1,2,1]) # The above logic does not work with repetation of elemets in the list
```

#### Number of permutations with repeated elements

```python
def get_permutations_repeatation(nums: list[int]) -> list[list[int]]:

    def backtrack():
        if len(currPermutation) == len(nums):
            permutations.append(currPermutation.copy())
            return

        # Determine which number should be placed in the current permutation list
        for num in num_count:
            if num_count[num] > 0:
                currPermutation.append(num)
                num_count[num] -= 1
                backtrack()

                currPermutation.pop()
                num_count[num] += 1

    permutations = []
    currPermutation = []
    
    # Count the occurrences of each number using a dictionary
    num_count = {}
    for num in nums:
        num_count[num] = num_count.get(num, 0) + 1
    
    backtrack()

    return permutations

# Test the function
print(get_permutations_repeatation([1, 2, 1]))
```
