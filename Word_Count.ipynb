```python
import re
from collections import Counter

# Define the dictionary for contractions
dict_map = {
    "i've": "i have",
    "he's": "he has",
    "she's": "she has",
    "it's": "it is",
    "i'm": "i am",
    "you're": "you are",
    "we're": "we are",
    "they're": "they are",
    "isn't": "is not",
    "aren't": "are not",
    "wasn't": "was not",
    "weren't": "were not",
    "can't": "cannot",
    "couldn't": "could not",
    "don't": "do not",
    "doesn't": "does not",
    "didn't": "did not",
    "won't": "will not",
    "wouldn't": "would not",
    "shouldn't": "should not",
    "hasn't": "has not",
    "haven't": "have not",
    "hadn't": "had not",
    "let's": "let us",
    "that's": "that is",
    "there's": "there is",
    "who's": "who is",
    "what's": "what is",
    "where's": "where is",
    "when's": "when is",
    "why's": "why is",
    "how's": "how is",
    "i'd": "i would",
    "you'd": "you would",
    "he'd": "he would",
    "she'd": "she would",
    "we'd": "we would",
    "they'd": "they would",
    "i'll": "i will",
    "you'll": "you will",
    "he'll": "he will",
    "she'll": "she will",
    "we'll": "we will",
    "they'll": "they will",
    "you've": "you have",
    "we've": "we have",
    "they've": "they have",
    "there'd": "there would",
    "it'd": "it would",
    "it'll": "it will",
    "who'll": "who will",
    "who'd": "who would",
    "what'll": "what will",
    "what'd": "what would"
}

def count_words(bow: str) -> dict:
    # Combine the patterns for full contractions and remaining cases
    pattern = re.compile(r'\b(' + '|'.join(re.escape(key) for key in dict_map.keys()) + r'|n\'t|\w+)\b', re.IGNORECASE)

    # Function to replace using the dictionary
    def replace_match(match):
        return dict_map.get(match.group(0).lower(), match.group(0))

    # Perform the replacements for contractions and preprocess the text
    bow = pattern.sub(replace_match, bow)
    bow = re.sub(r'\b[a-zA-Z]\b', '', bow)  # Remove single characters
    bow = re.sub(r'\s+', ' ', bow)  # Replace multiple spaces with single space
    bow = bow.strip().lower()  # Trim and convert to lowercase

    # Count the words using Counter
    word_count = Counter(bow.split())

    return word_count

# Example 1
text1 = "I've been there, he's seen it, she's done it, and it's amazing."
print(count_words(text1))
# Example 2
text2 = '''This isn't a simple test passage. 
It contains contractions like 'isn't', 'don't', and 'wouldn't'. 
There are also hyphenated words, such as 'well-known' and 'self-made'. 
Additionally, it includes numbers like '123' and special characters like '@#$%^&*'. 
We'll see how well the function handles these complexities!
'''
print(count_words(text2))
```
